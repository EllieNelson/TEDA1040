# Module 2 Project Melbourne Housing

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

### Question 1 
Create a scatter plot that shows land size on the x-axis and price on the y-axis. Change the color of the points to reflect the type of the property

```python
# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Landsize', y='Price', hue='Type', alpha=0.6)

# Add labels and title
plt.title('Property Price vs Land Size by Type')
plt.xlabel('Land Size')
plt.ylabel('Price')

# Format axis labels for better readability
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.3)

# Move legend to the outside of the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent label cutoff and accommodate legend
plt.tight_layout()

plt.show()
```

### Problem 3
Create a line chart that shows the year built along the x-axis and the average price per year on the y-axis. You will need to use aggregation and group by to accomplish this

```python
# Group by YearBuilt and calculate mean Price
yearly_avg_price = df.groupby('YearBuilt')['Price'].mean().reset_index()

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(yearly_avg_price['YearBuilt'], yearly_avg_price['Price'], marker='o')

# Add labels and title
plt.title('Average House Price by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Average Price')

# Format y-axis to show currency
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

```python
# Group by Regionname and calculate mean Price
region_avg_price = df.groupby('Regionname')['Price'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=region_avg_price.index, y=region_avg_price.values)

# Add labels and title
plt.title('Average Property Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Price')

# Format y-axis to show currency
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

```python
# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Price')

# Add labels and title
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Amount of homes')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.3)



# Format x-axis to show currency
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

### Question 9
Create a box plot that shows the YearBuilt that each property was built on the x-axis and the Regionname on the y-axis. These box plots should be oriented horizontally

```python
# Create the horizontal box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='YearBuilt', y='Regionname', orient='h')

# Add labels and title
plt.title('Distribution of Property Build Years by Region')
plt.xlabel('Year Built')
plt.ylabel('Region')

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.3)

# Format x-axis to show years without decimals
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x)}'))

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```
