```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

### Question 1
Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state.

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, x='state')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.title('Number of Companies by State')
plt.xlabel('State')
plt.ylabel('Count')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```

### Question 3
Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.

```python
# Group by industry and sum the job openings
industry_totals = df.groupby('Industry')['job_openings'].sum().sort_values(ascending=False)

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(y=industry_totals.values, x=industry_totals.index)

# Add labels and title
plt.title('Total Job Openings by Industry')
plt.ylabel('Number of Job Openings')
plt.xlabel('Industry')

# Adjust layout to prevent label cutoff
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.tight_layout()

# Display the plot
plt.show()
```

### Question 5 
Create a line chart that shows the average growth rate for each founded year.

```python
# Group by founded_year and calculate mean growth_percentage
yearly_growth = df.groupby('founded')['growth_percentage'].mean()

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(yearly_growth.index, yearly_growth.values, marker='o')

# Add labels and title
plt.title('Average Company Growth Percentage by Founding Year')
plt.xlabel('Founded Year')
plt.ylabel('Average Growth Percentage')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Format y-axis to show percentage
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.1f}%'))

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

# Question 7
Create a scatter plot that compares the number of employees to estimated revenues for each company.

```python
# Create the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='employees', y='estimated_revenues', alpha=0.5)

# Add labels and title
plt.title('Company Size vs Estimated Revenue')
plt.xlabel('Number of Employees')
plt.ylabel('Estimated Revenue')

# Format axis labels to be more readable (assuming revenues are in large numbers)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Will Not show some outliers but makes the plot more readable
plt.gca().set_ylim(0, 300000000)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

```python
sns.regplot(data=df, x='employees', y='estimated_revenues', scatter_kws={'alpha':0.5})
```

### Question 9
Create a histogram that measures the number of companies founded in each year

```python
# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='founded')

# Add labels and title
plt.title('Distribution of Company Founding Years')
plt.xlabel('Founded Year')
plt.ylabel('Number of Companies')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.3)

# Format y-axis to use comma separator for thousands
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'))

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```
